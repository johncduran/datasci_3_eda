```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)

outliers
```

Here this allowed me to see that there were outliers present in the dataset. Later on I will exlpore one of these outliers from line 25 of this dataset.

```python
numerical_columns = df.select_dtypes(include='number').columns
sns.pairplot(df[numerical_columns])
plt.show()
```

Here I was able to create scatter plots of all the numerical values in the data set. However, this isn't very easy on the eyes and so in the next step I will create only one scatter plot with two variables.

```python
# Define the two numerical variables you want to explore
x_var = 'Asthma'
y_var = 'Anxiety_Di'

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=x_var, y=y_var, color='skyblue')
plt.title(f'Scatter Plot between {x_var} and {y_var}')
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.show()
```

As you can see here I created a scatter plot between two numerical variables. Both 'Asthma' and 'Anxiety_Di' were both column names that only contained numerical values.

You can also tell that there are a few outliers in this exmaple at the far right of the plot.
